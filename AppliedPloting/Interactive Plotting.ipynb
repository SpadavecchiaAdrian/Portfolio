In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

## Generate data



```python
#Generating data

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

# Interactive Plot: 
Add interactivity to the plot, which allows the user to click on the y axis to set the value of interest. The bar colors should change with respect to what value the user has selected.

### data processing:


Due to the number of samples in the data, it is possible to perform the calculations more optimally, however I have left it to make the procedure more explicit.

```python

#transpose the data and getting basic features
desc = df.transpose().describe()

#95% confidence interval for the mean 
ci95_hi = desc.loc['mean'] + 1.96 * desc.loc['std']/np.sqrt(desc.loc['count'])
ci95_hi.name = 'ci95H'
desc = desc.append(ci95_hi)

ci95_lo = desc.loc['mean'] - 1.96 * desc.loc['std']/np.sqrt(desc.loc['count'])
ci95_lo.name = 'ci95L'
desc = desc.append(ci95_lo)

desc = desc.transpose()
desc
```

### Ploting:



```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable


#function to re draw when click envent
def plotbars(yvalue):
    
    #estimate the position normalizated 0-1 to map color
    NormPos = []
    for val in desc.index:
        if yvalue > desc.loc[val, 'ci95H']:
            NormPos.append(1.0)
        elif yvalue < desc.loc[val, 'ci95L']:
            NormPos.append(0.0)
        else:
            #value normalizated
            NormPos.append((yvalue - desc.loc[val, 'ci95L'])/(desc.loc[val, 'ci95H'] - desc.loc[val, 'ci95L']))
       

    #map colors
    colors = my_cmap(NormPos)

    #plot bars
    plt.bar(data_x, desc['mean'], width = 0.7 
            ,yerr= (desc['ci95H'] - desc['mean']) ,capsize = 8
            , edgecolor = 'black', color=colors)
    #line with y values
    plt.axhline(y=yvalue, color='black', linestyle='-')

    #set title
    plt.title('Click on the plot to select y value')
    plt.xticks(data_x)    
    plt.ylabel("Y")
        
    

#size plot    
plt.subplots(figsize=(9, 6))

data_x = [1992,1993,1994, 1995 ]


#barcolor
my_cmap = plt.cm.get_cmap('coolwarm')

sm = ScalarMappable(cmap=my_cmap)
sm.set_array([])

cbar = plt.colorbar(sm)
cbar.set_label('Confidence Interval')




# plot to dont get a empty frame
plotbars(41000)

#function to manage event click
def onclick(event):
    plt.cla() #clear axis
    plotbars(event.ydata)
    plt.gca().set_title('"Y" Value:  {0:.2f}'.format(event.ydata))


# tell mpl_connect we want to pass a 'button_press_event' into onclick when the event is detected
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```
